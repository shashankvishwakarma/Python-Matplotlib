```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.2f%%')
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.0f%%')
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.0f%%', shadow=True)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.0f%%', shadow=True, explode=[0,0.2,0,0,0])
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.0f%%', shadow=True, explode=[0,0.4,0,0,0])
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.0f%%', shadow=True, explode=[0,0.4,0,0,0], startangle=45)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.0f%%', shadow=True, explode=[0,0.4,0,0,0], startangle=180)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.0f%%', shadow=True, explode=[0,0.4,0,0,0], startangle=180)
plt.show()
plt.savefig("piechart.jpg")
```

# Saving chart to a file using savefig

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.0f%%', shadow=True, explode=[0,0.4,0,0,0], startangle=180)
# This should come after savefig or comment it if not required, else image will be empty
#plt.show() 
plt.savefig("piechart.jpg")
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.0f%%', shadow=True, explode=[0,0.4,0,0,0], startangle=180)
plt.savefig("piechart.jpg", bbox_inches="tight")
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.0f%%', shadow=True, explode=[0,0.4,0,0,0], startangle=180)
plt.savefig("piechart.jpg", bbox_inches="tight", pad_inches=2)
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.0f%%', shadow=True, explode=[0,0.4,0,0,0], startangle=180)
plt.savefig("piechart.jpg", bbox_inches="tight", pad_inches=2, transparent=True)
```

```python
plt.pie(exp_vals,labels=exp_labels, radius=1.5, autopct='%0.0f%%', shadow=True, explode=[0,0.4,0,0,0], startangle=180)
plt.savefig("piechart.pdf", bbox_inches="tight", pad_inches=2, transparent=True)
```
